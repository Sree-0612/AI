```python
import nltk
from nltk.corpus import wordnet
from gensim.models import Word2Vec

# Download WordNet data (if not already downloaded)
nltk.download('wordnet')

# Load pre-trained Word2Vec model
# Note: You need to download a pre-trained Word2Vec model or train your own on a large corpus
# For demonstration purposes, you can use a pre-trained model if available
# Example: https://github.com/RaRe-Technologies/gensim-data#models

w2v_model = Word2Vec.load("https://github.com/RaRe-Technologies/gensim-data#models")


def generate_synonyms(word, max_synonyms=10):
    synonyms = set()

    # Get synsets for the word
    synsets = wordnet.synsets(word)

    for synset in synsets:
        for lemma in synset.lemmas():
            synonym = lemma.name().replace('_', ' ')
            synonyms.add(synonym)
            if len(synonyms) >= max_synonyms:
                return list(synonyms)

    return list(synonyms)


def main():
    word = "UI/UX designer"  # Predefined word for demonstration
    max_synonyms = 10  # Maximum number of synonyms to generate

    # Generate synonyms using WordNet
    synonyms = generate_synonyms(word, max_synonyms)

    if synonyms:
        print(f"Synonyms for '{word}':")
        for synonym in synonyms:
            print("-", synonym)
    else:
        print(f"No synonyms found for '{word}'.")


if __name__ == "__main__":
    main()
```
