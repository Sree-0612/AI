```python
from transformers import pipeline

def generate_product_description(product_name):
    try:
        prompt = f"Write a detailed product description for {product_name}. Highlight its key features."
        generator = pipeline('text-generation', model='SFR-Embedding-2_R')
        response = generator(prompt, max_length=150, num_return_sequences=1, truncation=True)
        description = response[0]['generated_text'].strip()
        return description
    except Exception as e:
        return f"An error occurred: {e}"

product_name = input("Enter the product name: ")
description = generate_product_description(product_name)
print(f"Product Description:\n{description}")
```
