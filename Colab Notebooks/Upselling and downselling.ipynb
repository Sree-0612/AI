A sample test with 4 products of same brands

```python
# Sample data
products = {
    'Samsung Galaxy S9': 500,
    'Samsung Galaxy S10': 700,
    'Samsung Galaxy S20': 900,
    'Samsung Galaxy S21': 1100,
}

def recommend_products(products, current_product):
    # Sort products by price
    sorted_products = sorted(products.items(), key=lambda x: x[1])

    # Find the current product
    for i, product in enumerate(sorted_products):
        if product[0] == current_product:
            # Recommend the next more expensive product for upselling
            upsell = sorted_products[i + 1][0] if i + 1 < len(sorted_products) else None

            # Recommend the next less expensive product for downselling
            downsell = sorted_products[i - 1][0] if i - 1 >= 0 else None

            return upsell, downsell

# Test the function
current_product = 'Samsung Galaxy S10'
upsell, downsell = recommend_products(products, current_product)
print(f"For the product '{current_product}', the upsell recommendation is '{upsell}' and the downsell recommendation is '{downsell}'.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
df = pd.read_csv('/content/data.csv', encoding='latin-1')

# Ask the user for the product
product = input("Enter the product: ").strip().lower()

# Handle missing values in 'Description' column before clustering
df['Description'] = df['Description'].fillna('')  # Replace missing values with empty strings

# TF-IDF Vectorizer to convert text data to numerical values
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Description'])

# Calculate cosine similarity between user input and all products
user_product_vector = vectorizer.transform([product])
cosine_similarities = cosine_similarity(user_product_vector, X).flatten()

# Find the most similar products based on cosine similarity
similar_products_indices = cosine_similarities.argsort()[::-1]

# Print the top 50 similar products without repetition
printed_products = set()
count = 0
for index in similar_products_indices:
    row = df.iloc[index]
    if row['Description'] != product and row['Description'] not in printed_products:  # Exclude the input product itself and avoid repetition
        print(f"Product: {row['Description']}, Price: {row['UnitPrice']}")
        printed_products.add(row['Description'])
        count += 1
    if count == 50:  # Limit the number of printed products to 50
        break
```

#Zoho Products

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
df = pd.read_excel('/content/zoho.xlsx')

# Ask the user for the product
product = input("Enter the product: ").strip().lower()

# Handle missing values in 'Description' column before clustering
df['Description'] = df['Description'].fillna('')  # Replace missing values with empty strings

# Convert the 'Description' column to strings
df['Description'] = df['Description'].astype(str) #This will ensure all values are strings

# TF-IDF Vectorizer to convert text data to numerical values
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Description'])

# Calculate cosine similarity between user input and all products
user_product_vector = vectorizer.transform([product])
cosine_similarities = cosine_similarity(user_product_vector, X).flatten()

# Find and print the products above cosine similarity 0.4
similar_products = []
for index, sim in enumerate(cosine_similarities):
    if sim > 0.4:
        row = df.iloc[index]
        # Use the 'in' keyword to check if the product is in the description
        if product in row['Description'].lower():
            similar_products.append((row['Description'], row['UnitPrice']))

# Sort the similar products by price
similar_products.sort(key=lambda x: x[1])

# Find the upselling and downselling products
upselling_product = None
downselling_product = None

# Get the price of the entered product, handling the case where it's not found
product_price = df.loc[df['Description'].str.lower().str.contains(product), 'UnitPrice'].values
if product_price.size > 0:  # Check if the product was found
    product_price = product_price[0]
    for prod, price in similar_products:
        if price > product_price:
            upselling_product = (prod, price)
            break

    for prod, price in reversed(similar_products):
        if price < product_price:
            downselling_product = (prod, price)
            break
else:
    print(f"Product '{product}' not found in the dataset.")

# Print the upselling and downselling products
if upselling_product:
    print(f"Upselling product: {upselling_product[0]}, Price: {upselling_product[1]}")
else:
    print("There is no upselling product for this category.")

if downselling_product:
    print(f"Downselling product: {downselling_product[0]}, Price: {downselling_product[1]}")
else:
    print("There is no downselling product for this category.")
```
