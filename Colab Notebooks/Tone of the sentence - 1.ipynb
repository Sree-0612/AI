# Using TextBlob

```python
from textblob import TextBlob

def check_tone():
    # Get the sentence from the user
    sentence = input("Please enter a sentence: ")

    # Create a TextBlob object
    blob = TextBlob(sentence)

    # Get the polarity of the sentence
    polarity = blob.sentiment.polarity
    print(f"The polarity of the sentence is: {polarity}")

    # If the polarity is negative, print a message
    if polarity < 0:
        print("Use a appropriate tone to talk to me")
    else:
        print("Your tone is fine")

# Call the function
check_tone()
```

# Using VADER sentiment analysis tool from the nltk library

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')

def check_tone_vader():
    # Get the sentence from the user
    sentence = input("Please enter a sentence: ")

    # Create a SentimentIntensityAnalyzer object
    sia = SentimentIntensityAnalyzer()

    # Get the polarity scores of the sentence
    polarity_scores = sia.polarity_scores(sentence)

    # Print all the sentiment scores
    print(f"Sentiment scores - Positive: {polarity_scores['pos']}, Negative: {polarity_scores['neg']}, Neutral: {polarity_scores['neu']}, Compound: {polarity_scores['compound']}")

    # Determine sentiment based on the compound score
    if polarity_scores['compound'] < 0:
        print("Overall sentiment: Negative")
    else:
        print("Overall sentiment: Positive")

# Call the function
check_tone_vader()
```
