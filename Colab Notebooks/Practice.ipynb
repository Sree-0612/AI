```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('clips/mfaq')
```

```python
question = "How many candidates know java?"
answer_0 = "Im proficient in Java ,  C , C++ , jQuery, ajax "
answer_1 = "I love fruits"
answer_2 = "Python is my strength. Apart from that, I can work in java , C , C++ and Oracle"

query_embedding = model.encode(question)
corpus_embeddings = model.encode([answer_0, answer_1, answer_2])

print(util.semantic_search(query_embedding, corpus_embeddings))
```

```python
from transformers import pipeline
```

```python
qa_model = pipeline("question-answering")
question = "How many languages does the candidate know?"
context = "Im proficient in Java,C,C++,jQuery,ajax "
qa_model(question = question, context = context)
```
