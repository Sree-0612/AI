```python
import flair

def check_tone_flair():
    sentence = input("Please enter a sentence: ")

    sentence = flair.data.Sentence(sentence)

    classifier = flair.models.TextClassifier.load('sentiment')

    classifier.predict(sentence)

    sentiment_result = sentence.labels[0]

    print(f"The sentiment of the sentence is: {sentiment_result}")

    if 'NEGATIVE' in str(sentiment_result):
        print("Use a appropriate tone to talk to me")
    else:
        print("Your tone is fine")

check_tone_flair()
```
