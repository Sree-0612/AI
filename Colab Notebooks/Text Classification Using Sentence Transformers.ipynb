```python
import pandas as pd
import numpy as np
np.random.seed(2022)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
!unzip -q train.csv.zip -d train
```

```python
data = pd.read_csv("/content/train.csv.zip", engine="python")

```

```python
data_subset = data.head(2000)
```

```python
import spacy
import string
nlp = spacy.load("en_core_web_sm")
stop_words = nlp.Defaults.stop_words
print(stop_words)
```

```python
punctuations = string.punctuation
print(punctuations)
```

```python
def spacy_tokenizer(sentence):
    doc = nlp(sentence)
    mytokens = [word.lemma_.lower().strip() for word in doc]
    mytokens = [word for word in mytokens if word not in stop_words and word not in punctuations]
    return " ".join(mytokens)
```

```python
list_ = ["happy" , "apple"]
"-".join(list_)
```

```python
spacy_tokenizer("I am Eating")
```

```python
data_subset['tokenize'] = data_subset['comment_text'].apply(spacy_tokenizer)
```

```python
data_subset.head()
```

```python
data_subset['embeddings'] = data_subset['tokenize'].apply(model.encode)
```

```python
data_subset.head()
```

```python
X = data_subset['embeddings'].to_list()
y = data_subset['toxic'].to_list()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
from sklearn import metrics
predicted = LR.predict(X_test)
print("Logistic Regression Accuracy:",metrics.accuracy_score(y_test, predicted))
print("Logistic Regression Precision:",metrics.precision_score(y_test, predicted))
print("Logistic Regression Recall:",metrics.recall_score(y_test, predicted))
```
