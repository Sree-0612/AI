```python
import flair

def check_tone_flair():
    sentence_input = input("Please enter a sentence: ")

    # Load the sentence and perform sentiment analysis
    sentence = flair.data.Sentence(sentence_input)
    classifier = flair.models.TextClassifier.load('sentiment')
    classifier.predict(sentence)

    # Extract the sentiment result
    sentiment_result = sentence.labels[0]
    print(f"The sentiment of the sentence is: {sentiment_result}")

    # Calculate the opposite sentiment score
    if sentiment_result.value == 'POSITIVE':
        positive_score = sentiment_result.score
        negative_score = 1 - positive_score
        print(f"Negative Score: {negative_score}")
        print("Your tone is positive.")
    elif sentiment_result.value == 'NEGATIVE':
        negative_score = sentiment_result.score
        print(f"Positive Score (for negative sentence): {1 - negative_score}")
        print("Use an appropriate tone to talk.")
    else:
        print("Neutral sentence detected.")

check_tone_flair()
```

```python
from textblob import TextBlob

def analyze_sentiment(sentence):
    # Create a TextBlob object
    blob = TextBlob(sentence)

    # Analyze the sentiment of each word
    for word in blob.words:
        word_blob = TextBlob(word)
        print(f"The polarity of '{word}' is {word_blob.sentiment.polarity}")

    # Analyze the overall sentiment of the sentence
    print(f"\nThe overall sentiment of the sentence is {blob.sentiment.polarity}")

# Test the function with a sentence
analyze_sentiment("I'll get a chocolate if i fail the exam")
```

```python
import flair


def execute_task(task):
    print(f"Executing task: {task}")


def check_tone_and_execute_task(sentence):
    sentence_obj = flair.data.Sentence(sentence)

    classifier = flair.models.TextClassifier.load('en-sentiment')

    classifier.predict(sentence_obj)

    sentiment_results = [label.value for label in sentence_obj.labels]

    print(f"The sentiments of the sentence are: {sentiment_results}")

    positive_count = sentiment_results.count('POSITIVE') + sentiment_results.count('NEUTRAL')
    negative_count = sentiment_results.count('NEGATIVE')


    if negative_count > positive_count:
        print("Negative sentence")
    else:
        text_to_analyze = sentence.split(". ")[1:]
        positive_count = sum('POSITIVE' in sentiment for sentiment in sentiment_results)
        negative_count = sum('NEGATIVE' in sentiment for sentiment in sentiment_results)

        print(f"Positive count: {positive_count}")
        print(f"Negative count: {negative_count}")

        if negative_count > positive_count:
            print("Negative sentence")
        else:
            for text in text_to_analyze:
                task = text.split("to ", 1)[-1]
                execute_task(task)
```

# BERT

```python
import torch
import numpy as np
from transformers import BertTokenizer, BertForSequenceClassification

# Set random seed for reproducibility
seed = 0
torch.manual_seed(seed)
np.random.seed(seed)

# Load pre-trained BERT model and tokenizer fine-tuned for sentiment analysis
model_name = 'bert-base-uncased-finetuned-sst-2-english'
# Pass the access token here
tokenizer = BertTokenizer.from_pretrained(model_name, token='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')
model = BertForSequenceClassification.from_pretrained(model_name, token='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')
model.eval()

# Disable gradient calculations
torch.set_grad_enabled(False)

# Define the function to analyze sentiment
def analyze_sentiment(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs)
    logits = outputs.logits
    probabilities = torch.softmax(logits, dim=1).tolist()[0]

    if probabilities[1] > probabilities[0]:
        print("Positive sentence")
    else:
        print("Negative sentence")

# Example usage
sentence = "I love this movie, it's amazing!"
analyze_sentiment(sentence)
```

```python
import spacy

def check_tone_and_execute_task():
    nlp = spacy.load("en_core_web_sm")

    sentence = input("Please enter a sentence: ")

    doc = nlp(sentence)

    sentiment_score = sum([token.sentiment for token in doc])

    print(f"The sentiment score of the sentence is: {sentiment_score}")

    if sentiment_score < 0:
        print("Negative sentence")
    elif sentiment_score > 0:
        print("Positive sentence")


# Call the function
check_tone_and_execute_task()
```
